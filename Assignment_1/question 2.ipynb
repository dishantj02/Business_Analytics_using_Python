```python
#1st graph
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,8,2)
y = np.random.rand(4)

plt.plot(x,y)
plt.show()

```

```python
#2nd graph
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(0,2*np.pi,5000)
r = 4
x = r*np.sin(a)
y = r*np.cos(a)

plt.plot(x,y)
plt.show()
```

```python
#3rd graph
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(0,2*np.pi,5000)
x1=np.sin(3*a)
y1=np.cos(2*a)
x2=np.cos(3*a)
y2=np.sin(2*a)
plt.plot(x1,y1,x2,y2)

plt.show()
```
