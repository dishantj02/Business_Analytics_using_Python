```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/gdrive')
ps = pd.read_excel('/content/gdrive/MyDrive/Colab Notebooks/Regression Data.xlsx')
ps.head(10)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score 
from sklearn.model_selection import train_test_split
```

```python
x1=ps[['Volume']]
x2=ps[['Weight']]
x3=ps[['Weight','Volume']]
y=ps[['CO2']]
```

## **linear regression to predict CO2 emissions using volume**

```python
X_train, X_test, y_train, y_test = train_test_split(x1, y, test_size=0.4, random_state=45)
plt.scatter(X_train,y_train)
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
y_pred = lm.predict(X_test)
plt.scatter(X_test, y_test, color ='b')
plt.plot(X_test, y_pred, color ='k')
 
plt.show()
```

```python
print('R2square=',r2_score(y_test, y_pred))
```

# linear regression to predict CO2 emissions using weight




```python
X_train, X_test, y_train, y_test = train_test_split(x2, y, test_size=0.4, random_state=45)
plt.scatter(X_train,y_train)
```

# multiple regression to predict CO2 emission using both weight and
# volume



```python
X_train, X_test, y_train, y_test = train_test_split(x3, y, test_size=0.4, random_state=45)
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
y_pred = lm.predict(X_test)
plt.scatter(y_test, y_pred, color ='k')
```

```python
print('R2square=',r2_score(y_test, y_pred))
```
